```python
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
tips = sns.load_dataset('tips')
```

```python
tips
```

```python
#scatter plot -> axes level function
sns.scatterplot(data=tips, x='total_bill',y='tip',hue='sex',style='time',size='size')

#hue parameter decriminate the two categories...dono ko alg alg color me show krdeta h
#style parameter used to descriminate two columns
```

```python
#relplot -> figure level -> square shape  .........# scatter using relplot -> size and hue
sns.relplot(data=tips, x='total_bill',y='tip',kind='scatter',hue='sex',style='time',size='size')
```

```python
#kdeplot------ KDE can produce a plot that is less cluttered and more interpretable, especially when drawing multiple distributions.But it has the potential to introduce distortions if the underlying distribution is bounded or not smooth

# Rather than using discrete bins, a KDE plot smooths the observations with a Gaussian Kernel, producing a continuos density estimate
tips
```

```python
sns.kdeplot(data=tips,x='total_bill')
```

```python
# hue -> fill
sns.displot(data=tips,x='total_bill',kind='kde')
```

```python
# hue -> fill
sns.displot(data=tips,x='total_bill',kind='kde',hue='sex',fill=True)
```

```python
# Rugplot

# Plot marginal distributions by drawing ticks along the x and y axes.

# This function is intended to complement other plots by showing the location of individual observations in an unobtrusive way.
sns.kdeplot(data=tips,x='total_bill')
sns.rugplot(data=tips,x='total_bill')
```

```python
# Bivariate histogram
# A bivariate histogram bins the data within rectangles that tilt the plot
# and then shows the count of observations within each rectangle with the color
#sns.histplot(data=tips, x='total_bill', y='tip')
sns.displot(data=tips, x='total_bill', y='tip',kind='hist')
```

```python
#Bivariate Kdeplot
# a bivariate KDE plot smoothes the (x, y) observations with a 2D Gaussian
sns.kdeplot(data=tips, x='total_bill', y='tip')
```

## Matrix Plot

```python
# Heatmap

# Plot rectangular data as a color-encoded matrix
gap
```

```python
temp_df = gap.pivot(index='country',columns='year',values='lifeExp')

#axes level function
plt.figure(figsize=(15,15))
sns.heatmap(temp_df)
```

```python
# annot
temp_df = gap[gap['continent'] == 'Europe'].pivot(index='country',columns='year',values='lifeExp')

plt.figure(figsize=(15,15))
sns.heatmap(temp_df,annot=True,linewidth=0.5,cmap='viridis')
```

```python
# Clustermap

# Plot a matrix dataset as a hierarchically-clustered heatmap.

# This function requires scipy to be available

iris = px.data.iris()
```

```python
iris
```

```python
sns.clustermap(iris.iloc[:,[0,1,2,3]])
```
